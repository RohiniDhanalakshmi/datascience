QUAN&QUAL

```python
import pandas as pd
```

```python
dataset=pd.read_csv("Placement.csv")
```

```python
dataset.dtypes
```

```python
class univariate():
    def quanqual(dataset):
        quan=[]
        for number in dataset.columns:
            if(dataset[number].dtype=="float64"):
                quan.append(number)
            if(dataset[number].dtype=="int64"):
                quan.append(number)
        return quan
         
```

```python
quan=univariate.quanqual(dataset)
```

```python
quan
```

CENTRAL TENDENCY

```python
dataset.describe()
```

```python
des=pd.DataFrame(index=["mean","median","mode"],columns=quan)
```

```python
des
```

```python
for code in quan:
    des[code]["mean"]=dataset[code].mean()
    des[code]["median"]=dataset[code].median()
    des[code]["mode"]=dataset[code].mode()[0]
```

```python
des
```

Percentile

```python
des=pd.DataFrame(index=["mean","median","mode","Q1","Q2","Q3","Q4"],columns=quan)
```

```python
des
```

```python
for code in quan:
    des[code]["mean"]=dataset[code].mean()
    des[code]["median"]=dataset[code].median()
    des[code]["mode"]=dataset[code].mode()[0]
    des[code]["Q1"]=dataset.describe()[code]["25%"]
    des[code]["Q2"]=dataset.describe()[code]["50%"]
    des[code]["Q3"]=dataset.describe()[code]["75%"]
    des[code]["Q4"]=dataset.describe()[code]["max"]
```

```python
des
```

IQR

```python
des=pd.DataFrame(index=["mean","median","mode","Q1","Q2","Q3","Q4","IQR","lesser","greater"],columns=quan)
```

```python
des
```

```python
for code in quan:
    des[code]["mean"]=dataset[code].mean()
    des[code]["median"]=dataset[code].median()
    des[code]["mode"]=dataset[code].mode()[0]
    des[code]["Q1"]=dataset.describe()[code]["25%"]
    des[code]["Q2"]=dataset.describe()[code]["50%"]
    des[code]["Q3"]=dataset.describe()[code]["75%"]
    des[code]["Q4"]=dataset.describe()[code]["max"]
    des[code]["IQR"]=des[code]["Q3"]-des[code]["Q1"]
    des[code]["lesser"]=des[code]["Q1"]-1.5*des[code]["IQR"]
    des[code]["greater"]=des[code]["Q3"]+1.5*des[code]["IQR"]
```

```python
des
```

FINDING OUTLIER

```python
lesser=[]
greater=[]
for code in quan:
    if(dataset.describe()[code]["min"]<des[code]["lesser"]):
        #print(lesser)
        lesser.append(code)
    if(dataset.describe()[code]["max"]>des[code]["greater"]):
        #print(greater)
        greater.append(code)
    
```

```python
for i in lesser:
         dataset.describe()[i]["min"]<des[i]["lesser"]=des[i]["lesser"]
```
