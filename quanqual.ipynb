```python
quan=[]
qual=[]
for qq in data:
    if(data[qq].dtypes=="object"):
        #print(qual)
        qual.append(qq)
    else:
        #print(quan)
        quan.append(qq)
```

```python
des=pd.DataFrame(index=["mean","median","mode"],columns=quan)
for num in quan:
    #print(num)
    des.loc["mean",num]=data[num].mean()
    des.loc["median",num]=data[num].median()
    des.loc["mode",num]=data[num].mode()[0]
```

PERCENTILE

```python
des=pd.DataFrame(index=["mean","median","mode","Q1","Q2","Q3","Q4"],columns=quan)
for num in quan:
    #print(num)
    des.loc["mean",num]=data[num].mean()
    des.loc["median",num]=data[num].median()
    des.loc["mode",num]=data[num].mode()[0]
    des.loc["Q1",num]=data.describe()[num]["25%"]
    des.loc["Q2",num]=data.describe()[num]["50%"]
    des.loc["Q3",num]=data.describe()[num]["75%"]
    des.loc["Q4",num]=data.describe()[num]["max"]
```

```python
des
```

IQR

```python
des=pd.DataFrame(index=["mean","median","mode","Q1","Q2","Q3","Q4","iqr","1.5","lesser","greater","min","max2",],columns=quan)
for num in quan:
    #print(num)
    des.loc["mean",num]=data[num].mean()
    des.loc["median",num]=data[num].median()
    des.loc["mode",num]=data[num].mode()[0]
    des.loc["Q1",num]=data.describe()[num]["25%"]
    des.loc["Q2",num]=data.describe()[num]["50%"]
    des.loc["Q3",num]=data.describe()[num]["75%"]
    des.loc["Q4",num]=data.describe()[num]["max"]
    des.loc["iqr",num]=des.loc["Q3",num]-des.loc["Q1",num]
    des.loc["1.5",num]=1.5*des.loc["iqr",num]
    des.loc["lesser",num]=des.loc["Q1",num]-des.loc["1.5",num]
    des.loc["greater",num]=des.loc["Q3",num]+des.loc["1.5",num]
    des.loc["min",num]=data[num].min()
    des.loc["max2",num]=data[num].max()
```

```python
des
```

FINDING OUTLIER

```python
lesser=[]
greater=[]
for numb in quan:
    #print(num)
    if des.loc["lesser", numb]>des.loc["min",numb]:
            lesser.append(numb)
    if des.loc["greater", numb]<des.loc["max2",numb]:
            greater.append(numb)
```

```python
lesser
```

```python
greater
```

REPLACEING OUTLIER

```python
lesser = []
greater = []

for numb in lesser:
    if des.loc["lesser", numb] > des.loc["min", numb]:
        des.loc["lesser", numb] = des.loc["lesser", numb]
        lesser.append(numb)
    if des.loc["greater", numb] < des.loc["max2", numb]:
        des.loc["greater", numb] = des.loc["greater", numb]
        greater.append(numb)
```

```python
greater
```

```python
def frq(columns,data):
    frq=pd.DataFrame(columns=["uniquevalue","frequency","relativefrequency","cumulativefrequency"])
    frq["uniquevalue"]=data["ssc_p"].value_counts().index
    frq["frequency"]=data["ssc_p"].value_counts().values
    frq["relativefrequency"]=frq["frequency"]/103
    frq["cumulativefrequency"]=frq["relativefrequency"].cumsum()
    return frq
```

HISTOGRAM

```python
des=pd.DataFrame(index=["mean","median","mode","Q1","Q2","Q3","Q4","iqr","1.5","lesser","greater","min","max2","kurtos","skeweness"],columns=quan)
for num in quan:
    #print(num)
    des.loc["mean",num]=data[num].mean()
    des.loc["median",num]=data[num].median()
    des.loc["mode",num]=data[num].mode()[0]
    des.loc["Q1",num]=data.describe()[num]["25%"]
    des.loc["Q2",num]=data.describe()[num]["50%"]
    des.loc["Q3",num]=data.describe()[num]["75%"]
    des.loc["Q4",num]=data.describe()[num]["max"]
    des.loc["iqr",num]=des.loc["Q3",num]-des.loc["Q1",num]
    des.loc["1.5",num]=1.5*des.loc["iqr",num]
    des.loc["lesser",num]=des.loc["Q1",num]-des.loc["1.5",num]
    des.loc["greater",num]=des.loc["Q3",num]+des.loc["1.5",num]
    des.loc["min",num]=data[num].min()
    des.loc["max2",num]=data[num].max()
    des.loc["kurtos",num]=data[num].kurtosis()
    des.loc["skeweness",num]=data[num].skew()
```
