```python
import pandas as pd
```

```python
data=pd.read_csv("LoanDataset.csv")
```

```python
quan=[]
qual=[]
for pre in data.columns:
    #print(pre)
    if(data[pre].dtypes=="object"):
        #print(qual)
        qual.append(pre)
    else:
        #print(quan)
        quan.append(pre)
```

```python
import numpy as np
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan,strategy='mean')
imp.fit(data[quan])
transformed_data=imp.transform(data[quan])
transformed_df=pd.DataFrame(transformed_data,columns=quan)
transformed_df
```
